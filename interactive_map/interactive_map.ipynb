```python
#create map and set extent
m = folium.Map(location=[35, -120])
m.fit_bounds([sw, ne])

#add station markers
for i in range(n_stations):
    name = stations.name[i]
    lat = stations.lat[i]
    lon = stations.lon[i]
    since = stations.since[i]
    sst = stations.sst[i]
    sbt = stations.sbt[i]
    link = stations.link[i]
    img_src = "https://via.placeholder.com/200" # TODO unhardcode and replace with stations.img_src[i]
    img_path = 'thumbnails/'+str(i)+'.png'
    tl = tl #TODO unhardcode
    color = temp_scale(sst)

    marker = folium.CircleMarker(location = [lat,lon],
                                 radius = 1,
                                 color = color,
                                 fill=True,
                                 fill_color = color,
                                 fill_opacity = 1)
    
    tooltip = folium.Tooltip(name) #TODO add photo
    marker.add_child(tooltip)
    popup_html = '''
    <!DOCTYPE html>
        <html>
            <head>
                <title>''' + name + '''</title>
            </head
            <body>
                <b><a href="''' + link + '''" target="_blank">''' + name + '''</a></b>
                <p>''' + str(lat) + '''°N, ''' +str(abs(lon)) + '''°W <br>
                Current Surface Temp: ''' + str(sst) + '''ºC at ''' + str(tl) + '''<br>
                Current Bottom Temp: ''' + str(sbt) + '''ºC at ''' + str(tl) + '''<br>'''
            #TODO better datetime format
    popup = folium.Popup(IFrame(html=popup_html, width=400, height=200))
    marker.add_child(popup)
    marker.add_to(m)
m
```
